#Finding Solution Set 'x' .
for i = 1 to n 
    if total+w[i] ≤ W then  
        x[i] = 1 
        total = total + w[i] 
    else 
        x[i] = (W - total) / w[i]
        total = W 
        break 
return x

### Let us Consider Maximal Capacity be 500.

```python
W=500
total=profit=0
for i in range(0,len(data)):
    if (total+data['weight'].iloc[i] <= W):
        data['x'].iloc[i]=1
        total=total+data['weight'].iloc[i]
    else:
        data['x'].iloc[i]=(W-total)/data['weight'].iloc[i]
        total = W
        break
print(data)
```
