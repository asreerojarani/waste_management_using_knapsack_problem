#### Data - https://www.kaggle.com/tizianoparriani/kp01pisinger?select=knapPI_1_500_1000_1_items.csv

```python
#Loading the data
import pandas as pd
data=pd.read_csv("knapsack.csv")
data
```

```python
wt=(data['Weight']).to_numpy()
wt
```

```python
v=(data['Profit']).to_numpy()
v
```

### Let the Maximal Capacity(W) be 500.

```python
n=len(wt)
W=500
F = [[0 for i in range(0,n)] for j in range(0,W)] 
for i in range(1,n):
    for j in range(1,W):
            if (j-wt[i])<0:
                F[i][j]=F[i-1][j]
            elif (j-wt[i])>=0:
                F[i][j]=max(F[i-1][j],v[i]+F[i-1][j-wt[i]])
print('Maximum Profit is:',F[n-1][W-1])
```
